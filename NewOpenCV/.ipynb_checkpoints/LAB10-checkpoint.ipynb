```python
"""
Import bibliotek potrzebnych przy realizacji projektu
"""
import cv2
import numpy as np
import matplotlib.pyplot as plt
import skimage
```

```python
"""
Utworzenie klasy MORF zawierającej metody wykorzystywane przy realizacji zadania
"""


class MORF:
    """
    Konstrukotr klasy MORF, który przeprowadza konwersję obrazu do skali szarości oraz binaryzację obrazu wejściowego. Wyniki zapisywane są do pól obiektu danej klasy.
    """

    def __init__(self, source):
        self.baseImage = source
        self.grayImage = cv2.cvtColor(self.baseImage, cv2.COLOR_BGR2GRAY)
        (self.tmp_thresh, self.binImage) = cv2.threshold(self.grayImage, 0, 1, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    """
    Metoda przeprowadzająca operację otwarcia na wejściowym obrazie binarnym
    """

    def Open(self, source):
        operator = source.copy()
        kernel = np.ones((24, 24), np.uint8)
        eroded_operator = cv2.erode(operator, kernel, iterations=1)
        open_operator = cv2.dilate(eroded_operator, kernel, iterations=1)
        return open_operator

    """
    Metoda przeprowadzająca operację zamknięcia na wejściowym obrazie binarnym
    """

    def Close(self, source):
        operator = source.copy()
        kernel = np.ones((24, 24), np.uint8)
        dilated_operator = cv2.dilate(operator, kernel, iterations=1)
        closed_operator = cv2.erode(dilated_operator, kernel, iterations=1)
        return closed_operator

    """
    Metoda przeprowadzająca operację otwarcia właściwego na wejściowym obrazie binarnym
    """

    def OW(self, source):
        return np.minimum(source, self.Close(self.Open(self.Close(source))))

    """
    Metoda przeprowadzająca operację zamknięcia właściwego na wejściowym obrazie binarnym
    """

    def ZW(self, source):
        return np.maximum(source, self.Open(self.Close(self.Open(source))))

    """
    Metoda wyliczająca automedianę dla wejściowego obrazu binarnego
    """

    def Automediana(self, source):
        return np.maximum(self.Open(self.Close(self.Open(source))), self.OW(source))

    """
    Metoda wyliczająca gradient morfologiczny dla wejściowego obrazu binarnego
    """

    def GradientMorfologiczny(self, source):
        kernel = np.ones((3, 3), np.uint8)
        dil = cv2.dilate(source, kernel, iterations=1)
        ero = cv2.erode(source, kernel, iterations=1)
        result = dil - ero
        return result

    """
    Metoda wyliczająca dekompozycję wewnętrznego gradientu morfologicznego dla wejściowego obrazu binarnego
    """

    def GradWewn(self, source):
        kernel = np.ones((3, 3), np.uint8)
        ero = cv2.erode(source, kernel, iterations=1)
        result = source - ero
        return result

    """
    Metoda wyliczająca dekompozycję zewnętrznego gradientu morfologicznego dla wejściowego obrazu binarnego
    """

    def GradZewn(self, source):
        kernel = np.ones((3, 3), np.uint8)
        dil = cv2.dilate(source, kernel, iterations=1)
        result = dil - source
        return result

    """
    Metoda wyliczająca laplasjan morfologiczny dla wejściowego obrazu binarnego
    """

    def LaplasjanMorfologiczny(self, source):
        grad_out = self.GradZewn(source)
        grad_in = self.GradWewn(source)
        result = grad_out - grad_in
        return result

    """
    Metoda realizująca filtrację Top-Hat (white top-hat) dla wejściowego obrazu binarnego
    """

    def Top_Hat_Filter(self, source):
        open_image = self.Open(source)
        result = source - open_image
        return result

    """
    Metoda realizująca filtrację Bottom-Hat (black top-hat) dla wejściowego obrazu binarnego
    """

    def Bottom_Hat_Filter(self, source):
        closed_image = self.Close(source)
        result = closed_image - source
        return result

    """Metoda realizująca szkieletyzację wejściowego obrazu binarnego"""

    def Skeletonize(self, source):
        result = skimage.morphology.skeletonize(source, method='lee')
        return result

    """Metoda obliczająca centroidy dla wejściowego obrazu binarnego"""

    def Centroid(self, source):
        prev_skel = source.copy()
        centroid = source.copy()

        struct1 = np.array([[0, -1, -1], [0, 1, 1], [0, 0, 0]])
        struct2 = np.array([[-1, 0, -1], [0, 1, 1], [-1, 0, -1]])
        counter = 0

        while True:
            for i in range(1, prev_skel.shape[0] - 1):
                for j in range(1, prev_skel.shape[1] - 1):
                    sector = np.array([[prev_skel[i - 1, j - 1], prev_skel[i - 1, j], prev_skel[i - 1, j + 1]],
                                       [prev_skel[i, j - 1], prev_skel[i, j], prev_skel[i, j + 1]],
                                       [prev_skel[i + 1, j - 1], prev_skel[i + 1, j], prev_skel[i + 1, j + 1]]])
                    if counter % 2 == 0:
                        if sector[0, 0] == struct1[0, 0] and sector[1, 0] == struct1[1, 0] and sector[1, 1] == struct1[
                            1, 1] and sector[1, 2] == struct1[1, 2] and sector[2, 0] == struct1[2, 0] and sector[
                            2, 1] == struct1[2, 1] and sector[2, 2] == struct1[2, 2]:
                            centroid[i, j] = 0
                            counter += 1
                    if counter % 2 != 0:
                        if sector[0, 1] == struct2[0, 1] and sector[1, 0] == struct2[1, 0] and sector[1, 1] == struct2[
                            1, 1] and sector[1, 2] == struct2[1, 2] and sector[2, 1] == struct2[2, 1]:
                            centroid[i, j] = 0
                            counter += 1

            if np.array_equal(centroid, prev_skel):
                break
            else:
                prev_skel = centroid

        return centroid

    def bwulterode(self, source):
        pass

    def convexHull(self):
        result = self.baseImage.copy()
        contours, _ = cv2.findContours(self.binImage, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        for i in contours:
            hull = cv2.convexHull(i)
            cv2.drawContours(result, [hull], -1, (0, 255, 255), 3)
        return result

    def convexityDefects(self):
        result = self.baseImage.copy()
        contours, hierarchy = cv2.findContours(self.binImage, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        for i in contours:
            hull = cv2.convexHull(i, returnPoints=False)
            defects = cv2.convexityDefects(i, hull)
            for j in range(defects.shape[0]):
                s, e, f, d = defects[j, 0]
                start = tuple(i[s][0])
                end = tuple(i[e][0])
                far = tuple(i[f][0])
                cv2.line(result, start, end, [0, 255, 0], 2)
                cv2.circle(result, far, 5, [0, 0, 255], -1)
        return result
```
